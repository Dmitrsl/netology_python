## 4. Визуализировать несколько взаимосвязей между переменными

```python
corr_threshold = 0.1
main_corr = hr.corr().abs().unstack().sort_values(ascending=False).drop_duplicates()[1:]
main_corr = main_corr[main_corr > corr_threshold]
main_corr
```

```python
corr_threshold = 0.1
main_corr_ = hr.corr().abs().unstack().sort_values(ascending=False).drop_duplicates()[1:]
main_corr = main_corr[main_corr > corr_threshold]
main_corr
```

```python
for corr_ in main_corr.index:
    plt.figure(figsize=(15,8))
    sns.scatterplot(data=hr, x=corr_[0], y=corr_[1]).set(
       title=f"{corr_[0]}, {corr_[1]}, коэффициент корреляции: {main_corr[corr_]:.2f}")
    plt.show()

```

Рассматрим взаимосвязи признaков с коэффициентом корреляции больше 0.3
    1. Наиболее яркая взаимосвязь между среднемесячным количеством часов и количеством проектов. Взаимосвязь положительная, с увеличением количества проектов увеличивается количество часов. Коэффициент корреляции 0.42
    2. Отрицательная взаимосвязь между уровнем удовлетворенности и фактом смены места работы. Коэффициент корреляции -0.39
    3. 


```python
corr_columns = set(np.array(main_corr.index.levels).ravel())
corr_columns
```

## 5. С использованием t-test статистики проверить взаимосвязи между несколькими выборками (двумя или больше)
